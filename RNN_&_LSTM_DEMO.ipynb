```python
input= [1,2,3,4]
output=[2,4,6,8]
```

```python
w = 0.0
b = 0.0
lr = 0.01
```

```python
for epoch in range(100):
  total_loss=0
  for x,y in zip(input,output):
    y_pred=w*x+b
    error=y_pred-y
    loss = error**2
    total_loss = total_loss + loss

    dw= 2*error*x
    db= 2*error

    w= w-lr*dw
    b= b-lr*db

  if epoch%10==0:
    print(f"epoch:{epoch},   Loss:{total_loss},  w:{w},    b:{b}")
    print(" ")

test_data = 7
prediction = w*test_data + b
print(prediction)
```

```python
import tensorflow as tf
import numpy as np
```

```python
x = np.array(input,dtype=float)
y = np.array(output, dtype=float)
```

Simple RNN

```python
model = tf.keras.Sequential([tf.keras.layers.Dense(20,activation='relu',input_shape = [1]),
                    tf.keras.layers.Dense(1) ])
```

```python
model.compile(optimizer=tf.keras.optimizers.SGD(learning_rate=.01),loss='mean_squared_error')
```

```python
model.fit(x,y,epochs=1000,verbose=1)
model.save("euron_gen_ai_first.h5")
```

```python
import numpy as np
test_data = np.array([19])
model.predict(test_data)
```

LSTM

```python
data='need'
```

```python
char2idx = {'n':0 , 'e' : 1 , "e":2 , "d" :3}
idx2char = {v : k for k , v in char2idx.items()}
```

```python
input_seq = [char2idx[c] for c in data[:-1]]
target_seq = [char2idx[c] for c in data [1:]]
```

```python
input_seq
```

```python
x = tf.one_hot(input_seq,depth=4 )
x = tf.expand_dims(x , axis=0)
y = tf.expand_dims(target_seq , axis=0)
```

```python
x
```

```python
model = tf.keras.Sequential([tf.keras.layers.LSTM(8,return_sequences=True , input_shape =(3,4) ) ,
                     tf.keras.layers.Dense(4,activation='linear')])
```

```python
model.compile(optimizer='adam' ,loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True))
```

```python
for epoch in range(201):
    model.train_on_batch(x,y)

    if epoch % 50  == 0 :
        output = model(x)
        print(output)
        pred = tf.argmax(output,axis = -1).numpy()[0]
        pred_char = "".join([idx2char[i] for i in pred] )
        current_loss = model.evaluate(x, y, verbose=0)
        print(f"epoch : {epoch} , prediction : {pred_char} , loss : {current_loss} ")
```

```python
model.save('euron_lstm.h5')
```
